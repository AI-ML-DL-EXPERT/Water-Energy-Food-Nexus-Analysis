```python
df = df.groupby("State Name")[['RICE YIELD (Kg per ha)', 'WHEAT YIELD (Kg per ha)', 
                          'KHARIF SORGHUM YIELD (Kg per ha)', 'RABI SORGHUM YIELD (Kg per ha)',
                          'COTTON YIELD (Kg per ha)', 'FRUITS AND VEGETABLES AREA (1000 ha)',
                          'SUGARCANE YIELD (Kg per ha)']].mean()
```

```python
northernRegion = ["Haryana", "Punjab", "Uttar Pradesh", "Uttarakhand", "Himachal Pradesh"]
centralRegion = ["Chhattisgarh", "Madhya Pradesh"]
southernRegion = ["Andhra Pradesh", "Karnataka", "Kerala", "Tamil Nadu", "Telangana"]
easternRegion = ["Assam", "Bihar", "Jharkhand", "Orissa", "West Bengal"]
westernRegion = ["Gujarat", "Maharashtra", "Rajasthan"]

region_dict = {
    "Northern Region": northernRegion,
    "Central Region": centralRegion,
    "Southern Region": southernRegion,
    "Eastern Region": easternRegion,
    "Western Region": westernRegion
}
```

```python
df.reset_index(inplace= True)
```

```python
df
```

```python
# Renaming the state name's with regions
def indianRegions(stateNameDF):
    for key, value in region_dict.items():
        for stateLst in value:
            if  stateNameDF == stateLst:
                return key

df["State Name"] = df["State Name"].apply(indianRegions)
```
