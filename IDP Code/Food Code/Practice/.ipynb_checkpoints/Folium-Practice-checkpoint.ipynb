```python
import folium
```

```python
m = folium.Map([26.763997876346085, 83.37506796835494])
m
```

```python
m = folium.Map(location= [26.763997876346085, 83.37506796835494], zoom_start=8)
m
```

```python
folium.Marker(location=[26.75081579617607, 83.4423592281201], popup= "Gorakhpur Airport").add_to(m)

m
```

```python
folium.Marker(location=[26.75081579617607, 83.4423592281201], 
              popup= "Home",
              icon= folium.Icon(icon= "home")).add_to(m)

m
```

```python
m.add_child(folium.ClickForMarker(popup="Places I want to explore!"))

m
```

```python
m.add_child(folium.LatLngPopup())

m
```

```python
m = folium.Map(location= [26.763997876346085, 83.37506796835494], zoom_start=14)

folium.Circle(
    radius= 1000,
    color = "blue",
    location= [26.757430845012212, 83.40940476406247],
    fill= True
).add_to(m)

folium.CircleMarker(
    radius= 50,
    color = "red",
    location= [26.752621914041004, 83.39359495038396],
    fill= False
).add_to(m)


m
```
